```python
x = np.linspace(start = 0, stop = 10, num = 50)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 0.5])
ax2 = fig.add_axes([0, 0.7, 0.8, 0.4])

ax1.plot(x, np.sin(x))

ax2.plot(x, np.cos(x))

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 0.5])

ax2 = fig.add_axes([0, 0.7, 0.8, 0.4])

ax1.plot(x, np.sin(x))
ax1.set_xlabel('x', fontsize = 18, color = 'g')
ax1.set_ylabel('sin(x)', fontsize = 15, color = 'm')

ax2.plot(x, np.cos(x))
ax2.set_xlabel('x', fontsize = 18, color = 'g')
ax2.set_ylabel('cos(x)', fontsize = 15, color = 'm')

plt.show()
```

```python
ax1 = plt.subplot2grid((2, 3), (0, 0))
ax1.plot(x, np.cos(x))

ax2 = plt.subplot2grid((2, 3), (0, 1))
ax2.plot(x, np.sin(x))

ax3 = plt.subplot2grid((2, 3), (0, 2), rowspan = 2)
ax3.plot([2, 3, 4, 5],
         [8, 6, 4, 2]         
        )
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2, 3), (1, 0), colspan = 2)
ax4.plot(x, np.exp2(x))

plt.show()
```

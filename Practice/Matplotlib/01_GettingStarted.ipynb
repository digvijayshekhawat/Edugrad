```python
x = np.linspace(start = -4, stop = 4, num = 256)
```

```python
plt.plot(x, np.sin(x))

plt.xlabel('x', fontsize = 15, color = 'r')
plt.ylabel('sin(x)', fontsize = 20, color = 'r')

plt.show()
```

```python
plt.tick_params(axis = 'y',
                color = 'red',
                labelcolor = 'blue',
                labelsize = 'xx-large')

plt.tick_params(axis = 'x',
                bottom = False,
                labelbottom = False)

plt.plot(x, np.sin(x))

plt.xlabel('x', fontsize = 15, color = 'r')
plt.ylabel('sin(x)', fontsize = 20, color = 'r')

plt.show()
```

```python
plt.plot(x, np.sin(x), label = 'sine curve')

plt.xlabel('x', fontsize = 18, color = 'm')
plt.ylabel('sin(x)', fontsize = 15, color = 'b')

plt.legend()

plt.show()
```

```python
plt.plot(x, np.sin(x), label = 'sine curve')

plt.xlabel('x', fontsize = 18, color = 'm')
plt.ylabel('sin(x)', fontsize = 15, color = 'b')

plt.legend()
plt.title('Play Plot Learn')

plt.xlim(2, 6)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
import matplotlib
matplotlib.is_interactive()
```

```python
x = np.linspace(start = 0, stop = 7, num = 256)
```

```python
plt.plot(x, np.sin(x))

plt.show()
```

```python
plt.plot(x, np.sin(x), label = 'sine curve')
plt.plot(x, np.cos(x), label = 'cosine curve')

plt.legend()
plt.title('Play Plot Learn')

plt.show()
```

```python
plt.plot(x, np.sin(x), label = 'sine curve', color = 'purple')
plt.plot(x, np.cos(x), label = 'cosine curve', color = 'g')

plt.legend()
plt.title('Play Plot Learn')

plt.show()
```

```python
random_array = np.random.randn(15)
```

```python
plt.plot(random_array, color = 'magenta')
plt.show()
```

```python
plt.plot(random_array, 
         color = 'orange',
         linestyle = ':')
plt.show()
```

```python
plt.plot(random_array, 
         color = 'cyan',
         linestyle = '--')
plt.show()
```

```python
plt.plot(random_array, 
         color = 'm',
         linestyle = '--',
         linewidth = 3)
plt.show()
```

```python
plt.plot(random_array,
         color = 'm', 
         marker = 'd')
plt.show()
```

```python
plt.plot(random_array,
         color = 'm', 
         marker = 'd',
         markersize = 10)
plt.show()
```

```python
plt.plot(random_array,
         color = 'm', 
         marker = 'd',
         linestyle = 'None')
plt.show()
```

```python
plt.scatter(range(0, 15),
            random_array,
            color = 'm', 
            marker = 'd')
plt.show()
```
